```python
df = pd.read_excel('/content/Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Flitro por objeto
filtro1 = df[df['PRECIO_UNITARIO'] == 140]
filtro1
```

```python
#Filtro por filas
filtro2 = df.iloc[5:15,:]
filtro2
```

```python
#Filtro por columnas no consecutivas
filtro3 = df.iloc[ : , [1,4,6]]
filtro3
```

```python
df1 = pd.read_excel('/content/Detalle precios y productos fabricados 2022.xlsx', index_col = 3)
df1
```

```python
filtro7 = df[df['PRECIO_UNITARIO'] > 500]
filtro7
```

```python
#Filtro and
filtro8=df[(df["PRECIO_UNITARIO"] > 500) & (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro8
```

```python
#Filtro "OR"
filtro9=df[(df["COSTO_UNITARIO"] > 1000) | (df["COSTO_UNITARIO"] > 500)]
filtro9
```

```python
#Filtro not
filtro10=df[~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL") & ~(df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro10
```
